```python
credit_renamed=credit.rename(index=str,columns={'movie_id':'id'})
credit_renamed.head()
```

```python
merge=movies.merge(credit_renamed,on='id')
merge.head()
```

```python
cleaned=merge.drop(columns=['homepage','title_x','title_y','status','production_countries'])
cleaned.head()
```

```python
cleaned['overview'].head()
```

```python
#将影评信息进行处理
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english',ngram_range=(1,3),min_df=3,analyzer='word')  #去除停顿词
cleaned['overview'] = cleaned['overview'].fillna('')  #去除空白影评
tfidf_matrix = tfidf.fit_transform(cleaned['overview']) #矩阵转置
tfidf_matrix.shape
```

```python
#计算关联矩阵
from sklearn.metrics.pairwise import linear_kernel
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
cosine_sim.shape
```

```python
indices = pd.Series(cleaned.index, index=cleaned['original_title']).drop_duplicates()
```

```python
#获取推荐影片
def get_recommendations(title, cosine_sim=cosine_sim):
    # 寻找影片名
    idx = indices[title]

    # G获取相似矩阵得分匹配
    sim_scores = list(enumerate(cosine_sim[idx]))

    # 筛选
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # 取相似矩阵得分前十的信息
    sim_scores = sim_scores[1:11]

    # 获得影片名
    movie_indices = [i[0] for i in sim_scores]

    # 输出10个相似影片
    return cleaned['original_title'].iloc[movie_indices]
```
